### LOAN DEFAULT RISK ANALYSIS USING CUSTOMER FINANCIAL DATA

### PROBLEM STATEMENT

Financial institutions face significant risk when approving loans, as some borrowers may fail to repay their loans on time, leading to loan defaults.
Without proper analysis of customer financial and personal data, banks may struggle to identify high-risk applicants, resulting in increased losses and poor credit decisions.
The goal of this project is to analyze the loan dataset to understand borrower characteristics and identify key factors that influence loan approval and default risk.

### Solutions

Analyze customer demographic and financial data to understand borrower profiles

Identify patterns between income, loan amount, credit history, and loan status

Determine factors that contribute to loan approval and loan default risk

Support banks in making data-driven loan approval decisions

Reduce financial risk by identifying potentially high-risk borrowers early

### Approach

Data Understanding

Reviewed dataset features such as applicant income, loan amount, loan term, credit history, property area, and loan status.

Data Cleaning

Handled missing values in income, loan amount, and credit history

Converted categorical variables into usable formats

Removed inconsistent or duplicate records

Exploratory Data Analysis (EDA)

Analyzed loan approval rates across income levels

Studied the impact of credit history on loan approval

Compared loan status with property area, education

Feature Analysis

Evaluated relationships between applicant income and loan amount

Examined how credit history influences loan approval decisions

Visualization

Bar charts for loan approval vs rejection

Histograms for income and loan amount distribution

Comparative charts for credit history vs loan status

### Conclusions

Credit history plays a crucial role in loan approval decisions

Applicants with stable income and good credit history have higher approval rates

Loan amount and applicant income are strongly related to loan outcomes

Certain demographic and property factors influence loan approval trends

Data-driven analysis can help financial institutions reduce default risk and improve lending strategies

The insight can help banks evaluate approval patterns and policy effectiveness.

### PIE CHART-LOAN STATUS DISTRIBUTION

```python
import matplotlib.pyplot as plt

# Count loan status
loan_counts = df['loan_status'].value_counts()

# Pie chart
plt.figure()
plt.pie(
    loan_counts,
    labels=loan_counts.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Loan Status Distribution")
plt.show()
```

 know you read to loan_counts = df['loan_status'].value_counts()

Example:
0 â†’ Loan Not Approved count

2. plt.figure()
Creates a new figure (plot window) for the pie chart.
3. plt.pie(
Starts creating the pie chart.
a) loan_counts
Values used to determine the size of each slice.
b) labels=loan_counts.index
Uses loan status values (0 and 1) as labels on the pie chart.
c) autopct='%1.1f%%'
Displays the percentage of each category on the pie chart.
Example: 77.8%, 22.2%.
d) startangle=90
Rotates the pie chart so it starts at 90 degrees.
Makes the chart look more balanced and readable.
4. plt.title('Loan Status Distribution')
